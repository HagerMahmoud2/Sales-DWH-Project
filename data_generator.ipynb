```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
```

```python
# CREATE PRODUCTS_TABLE
def generate_products_data(n):
    products=[]
    for i in range(n):
        products.append({
            "product_id" : i+1,
            "product_name" : fake.word(),
            "catagory" : random.choice(["Electronics","Clothing","Fruniture"]),
            "brand" : fake.company(),
            "size" : random.choice(["Small","Medium","Large"]),
            "color" : random.choice(["Red","Blue","Green"]),
            "price" : round(random.uniform(100,5000),2)
        })
    return pd.DataFrame(products)
        
```

```python
products_df=generate_products_data(100)
```

```python
# CREATE STORE TABLE
def generate_stores_data(n):
    stores=[]
    for i in range(n):
        stores.append({
            "store_id" : i+1,
            "store_name" : fake.word(),
            "region" : fake.state(),
            "city" : fake.city(),
            "manager" : fake.name()

        })
    return pd.DataFrame(stores)
            
```

```python
stores_df = generate_stores_data(20)
```

```python
#CREATE CUSTOMERS TABLE
def generate_customer_data(n):
    customers=[]
    for i in range(n):
        
        customers.append({
                "customer_id" : i + 1,
                "name" : fake.name(),
                "gender" : random.choice(['Male',"Female"]),
                "age" : random.randint(18,65),
                "Occupation": fake.job(),
                "IncomeLevel": random.choice(['Low', 'Medium', 'High']),
                "City": fake.city()
            
        })
    return pd.DataFrame(customers)
```

```python
customers_df = generate_customer_data(200)
```

```python
#CREATE DATES TABLE
def generate_dates(n):
    dates = []
    for i in range(1, n + 1):
        date = fake.date_between(start_date='-1y', end_date='today')
        dates.append({
            "date_id": i,
            "Date": date,
            "Year": date.year,
            "Quarter": (date.month - 1) // 3 + 1,
            "Month": date.month,
            "Week": date.isocalendar()[1],
            "Day": date.day,
            "Period": "AM" if random.choice([True, False]) else "PM"
        })
    return pd.DataFrame(dates)
```

```python
dates_df = generate_dates(365)
```

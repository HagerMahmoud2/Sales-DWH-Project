```python
from faker import Faker
import pandas as pd
import random
```

```python
fake = Faker()
```

```python
# CREATE PRODUCTS_TABLE
def generate_products_data(n):
    products=[]
    for i in range(n):
        products.append({
            "product_id" : fake.uuid4(),
            "product_name" : fake.word(),
            "catagory" : random.choice(["Electronics","Clothing","Fruniture"]),
            "brand" : fake.company(),
            "size" : random.choice(["Small","Medium","Large"]),
            "color" : random.choice(["Red","Blue","Green"]),
            "price" : round(random.uniform(100,5000),2)
        })
        return pd.DataFrame(products)
        
```

```python
df_products=generate_products_data(100)
```

```python
# CREATE STORE TABLE
def generate_stores_data(n):
    stores=[]
    for i in range(n):
        stores.append({
            "store_id" : fake.uuid4(),
            "store_name" : fake.word(),
            "region" : fake.state(),
            "city" : fake.city(),
            "manager" : fake.name()

        })
        return pd.DataFrame(stores)
            
```

```python
df_stores = generate_stores_data(20)
```

```python
#CREATE CUSTOMERS TABLE
def generate_customer_data(n):
    customers=[]
    for i in range(n):
        
        customers.append({
                "customer_id" : fake.uuid4(),
                "name" : fake.name(),
                "gender" : random.choice(['Male',"Female"]),
                "age" : random.randint(18,65),
                "Occupation": fake.job(),
                "IncomeLevel": random.choice(['Low', 'Medium', 'High']),
                "City": fake.city()
            
        })
    return pd.DataFrame(customers)
```

```python
df_customers = generate_customer_data(200)
```
